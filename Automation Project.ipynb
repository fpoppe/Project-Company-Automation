### 1. Importing the Data Bases

```python
import pandas as pd

#lendo os dataframes
emails = pd.read_excel(r'Bases de Dados\Emails.xlsx')
stores = pd.read_csv(r'Bases de Dados\Lojas.csv',sep = ';', encoding = 'latin1')
sales = pd.read_excel(r'Bases de Dados\Vendas.xlsx')
display(emails,stores,sales)
```

### 2. Create Store dataframes and KPI current day

```python
from pathlib import Path

#juntando o dataframe de vendas com o ID lojas para termos os nomes das lojas com suas vendas
sales = sales.merge(stores, on = 'ID Loja')
display(sales)
```

```python
#criar um dataframe por loja e armazenar isso em um dicionário
dic_stores = {}
for loja in stores['Loja']:
    dic_stores[loja] = sales[sales['Loja'] == loja]
```

```python
#a partir da ultima data da base, armazenar a data mais recente
kpi_date = sales['Data'].max()
print(kpi_date)
```

### 3. Creating folders and saving the arquives on folders

```python
#verificando se pasta da loja já existe. Caso não exista, criá-la. Independente do cenário, adicionar a base na pasta

#caminho das pastas
folder_backup_path = Path(r'Backup Arquivos Lojas')

#navegar pelas pastas do caminho
folder_backup = folder_backup_path.iterdir()

#criar uma lista com os nomes das pastas no caminho
list_files = []
list_files = [files.name for files in folder_backup]

#para cada loja, criar a pasta, caso não exista
for store in dic_stores:
    if not (store in list_files):
        Path(r'Backup Arquivos Lojas/{}'.format(store)).mkdir()
    dic_stores[store].to_excel(r'Backup Arquivos Lojas\{}\{}_{}_{}_{}.xlsx'.format(store,store,kpi_date.day,kpi_date.month,kpi_date.year))
```

### 4. Calculating KPIs - Revenue, Average Ticker and Diversity of Products

```python
#kpi goals
goal_rev_day = 1000
goal_rev_year = 1650000
goal_avrgT_day = 500
goal_avrgT_year = 500
goal_pdiversity_day = 4
goal_pdiversity_year = 120
```

```python
loja = 'Salvador Shopping'
sales_store = dic_stores[loja]
sales_store_day = sales_store.loc[sales_store['Data'] == kpi_date,:]

#Revenue
revenue_day = sales_store_day['Valor Final'].sum()
revenue_year = sales_store['Valor Final'].sum()

#Product Diversity
pdiversity_day = len(sales_store_day['Produto'].unique())
pdiversity_year = len(sales_store['Produto'].unique())

#Average Ticket
value_sales = sales_store.groupby('Código Venda').sum()
value_sales_day = sales_store_day.groupby('Código Venda').sum()
avrg_ticket_day = value_sales_day['Valor Final'].mean()
avrg_ticket_year = value_sales['Valor Final'].mean()
```
